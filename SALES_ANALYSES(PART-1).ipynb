#Azo Insulated Lunch box has been sold highest as per quantity and salings

```python
prices = df4.groupby('Product').mean()['Amount']
product_group = df4.groupby('Product')
quantity_ordered = product_group.sum()['Qty']
keys = [pair for pair, df in product_group]

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(keys, quantity_ordered, color='g')
ax2.plot(keys, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Qty', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(keys, rotation='vertical', size=8)

fig.show()
```
