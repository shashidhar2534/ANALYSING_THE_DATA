```python
df_grouped = df4[['Order Date','Product', 'store_branch','Qty']]
df_grouped = df_grouped.groupby(['Order Date','Product', 'store_branch'], as_index=False)

# Grouped gives us the indices we want for each grouping
# We cannot convert a groupedby object back to a dataframe, so we need to do it manually
# Create a new dataframe to work against
df_aggregated = df_grouped.size().reset_index()




def manualAggregations(indices_array):
    temp_df = df4.iloc[indices_array]
    return {
        'total_qty': temp_df['Qty'].sum(),
        
    }

for name, group in df_grouped:
    ix = df_grouped.indices[name]
    calcDict = manualAggregations(ix)

    for key in calcDict:
        #Salary Basis, Job Title
        columns = list(name)
        df_aggregated.loc[(df_aggregated['Order Date'] == columns[0]) & 
                          (df_aggregated['Product'] == columns[1]) & (df_aggregated['store_branch'] == columns[2]) , key] = calcDict[key]
```

```python
df_aggregated
```

```python
df_aggregated['total_qty']=df_aggregated['total_qty'].astype('int32')
```

```python
df7=df_aggregated.drop(['index','size'],axis='columns')
```

```python
df7.to_csv("final_data.csv",index=False)
```

```python
X = df10.drop(['total_qty'],axis='columns')
X.head(3)
```

```python
y = df10.total_qty
y.head(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
model_layer = keras.Sequential([
    keras.layers.Dense(1000, input_shape=(1518,),activation='tanh' ),
    keras.layers.Dense(1873,activation='sigmoid')
])

model_layer.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model_layer.fit(X_train,y_train, epochs=5)
```

```python
model_layer.evaluate(X_test,y_test)
```

```python
def predict_quantity(storebranch,product,month): 
    storebranch_index = np.where(X.columns==storebranch)[0][0]
    product_index = np.where(X.columns==product)[0][0]
    

    x = np.zeros(len(X.columns))
    x[0] = month
    
    if storebranch_index >= 0:
        x[storebranch_index] = 1
    if product_index >= 0:
        x[product_index] = 1
    arr=np.array([x])
    return np.argmax((model_layer.predict([arr][0])))
```
